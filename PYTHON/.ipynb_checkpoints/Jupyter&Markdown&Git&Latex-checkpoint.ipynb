##  Introduccion

El Proyecto Jupyter es una organización sin ánimo de lucro creada para "desarrollar software de código abierto, estándares abiertos y servicios para computación interactiva en docenas de lenguajes de programación". Creado a partir de IPython en 2014 por Fernando Pérez, el proyecto Jupyter soporta entornos de ejecución en varias docenas de lenguajes de programación. El nombre del proyecto Jupyter es una referencia a los tres lenguajes de programación principales soportados por Jupyter, que son Julia, Python y R, y también un homenaje a los cuadernos de Galileo que registran el descubrimiento de los satélites de Júpiter. El proyecto Jupyter ha desarrollado y respaldado los productos de computación interactiva Jupyter Notebook, JupyterHub y JupyterLab, la versión de próxima generación de Jupyter Notebook.

Jupyter Notebook (anteriormente IPython Notebooks) es un entorno informático interactivo basado en la web para crear documentos de Jupyter notebook. El término "notebook" puede hacer referencia coloquialmente a muchas entidades diferentes, principalmente la aplicación web Jupyter, el servidor web Jupyter Python o el formato de documento Jupyter según el contexto. Un documento de Jupyter Notebook es un documento JSON, que sigue un esquema versionado y que contiene una lista ordenada de celdas de entrada/salida que pueden contener código, texto (usando Markdown), matemáticas, gráficos y texto enriquecidos, generalmente terminado con la extensión ".ipynb".

Un Jupyter Notebook se puede convertir a varios formatos de salida estándar abierto (HTML, diapositiva de presentación, LaTeX, PDF, ReStructuredText , Markdown, Python) a través de "descargar como" en la interfaz web, a través de la biblioteca nbconvert o el comando "jupyter nbconvert" en un shell.

Para simplificar la visualización de los documentos Jupyter Notebook en la web se proporciona la biblioteca nbconvert como un servicio a través de NbViewer que puede llevar una URL a cualquier Notebook disponible al público, convertirlo a HTML sobre la marcha y mostrarlo al usuario.

El Jupyter Notebook se ha convertido en una interfaz de usuario popular para la computación en la nube, y los principales proveedores de la nube han adoptado el Jupyter Notebook o las herramientas derivadas como una interfaz front-end para los usuarios de la nube. Los ejemplos incluyen los SageMaker notebook de Amazon y Azure Notebook de Microsoft

https://jupyter.org/index.html

Markdown es un lenguaje de marcado ligero creado por John Gruber que trata de conseguir la máxima legibilidad y facilidad de publicación tanto en su forma de entrada como de salida, inspirándose en muchas convenciones existentes para marcar mensajes de correo electrónico usando texto plano. Se distribuye bajo licencia BSD y se distribuye como plugin (o al menos está disponible) en diferentes sistemas de gestión de contenidos (CMS). Markdown convierte el texto marcado en documentos XHTML utilizando html2text creado por Aaron Swartz. Markdown fue implementado originariamente en Perl por Gruber, pero desde entonces ha sido traducido a multitud de lenguajes de programación, incluyendo PHP, Python, Ruby, Java y Common Lisp.

## Listas desordenadas

Las listas desordenadas se elaboran antecediendo al texto de -, * o +.

Veamos su forma de escribir:

    - Elemento uno
    * Elemento dos
    + Elemento tres

- Elemento uno
* Elemento dos
+ Elemento tres

<a name=item5></a>
## Listas numeradas

Las listas numeradas se elaboran antecediendo al texto el numero correspondiente 1,2,3... etc. seguido de punto y espacio.

Veamos su forma de escribir:

    1. Elemento uno
    2. Elemento dos
    3. Elemento tres

1. Elemento uno
2. Elemento dos
3. Elemento tres

Para anidar las listas, debajo del item principal aplicamos tabuladores seguido de la etiqueta y el nombre del item.

Veamos su forma de escribir:

    1. Elemento uno
        1. Elemento uno-uno
        2. Elemento uno-dos
        3. Elemento uno-tres
    2. Elemento dos
    3. Elemento tres

1. Elemento uno
    1. Elemento uno-uno
    2. Elemento uno-dos
    3. Elemento uno-tres
2. Elemento dos
3. Elemento tres

## Negrilla, cursiva, subrayado y tachado

Para establecer texto en negilla es necesario encerrar el texto entre doble *.

La forma de escribirlo es de la siguiente manera:

    **Texto**

**Texto en Negrilla**

En HTML es:
    
    <b>Texto</b>

<b>Texto en Negrilla con HTML</b>

Para establecer texto en cursiva es necesario encerrar el texto entre *.

La forma de escribirlo es de la siguiente manera:

    *Texto*

*Cursiva*

Negrilla + cursiva se realiza con 3 *.
    
    ***Texto en negrilla y cursiva***

***Texto en Negrilla y Cursiva*** 

Para subrayar el texto es necesario utilizar la etiqueta de HTML < u >.

Su forma de escribirlo es asi:

    <u>Texto a subrayar</u>

<u>Texto subrayado</u>

Para tachar el texto es necesario encerrar el texto entre doble virgulilla (~ ~).

Su forma de escribirlo es asi:

    ~~Texto a tachar~~ 

Para tachar el texto es necesario encerrar el texto entre doble virgulilla (~ ~).

Su forma de escribirlo es asi:

    ~~Texto a tachar~~ 

~~Texto Tachado~~

## Enlaces

Para crear enlaces debes nada mas añadir entre corchetes la frase o palabra que quieras utilizar para anclar al enlace y despues de esto entre parentesis la URL del sitio al cual quieres que te redirija.

Su forma de escribirse es asi:

    [Frase que va anclar el enlace](http:Enlace.com)

[Este es un enlace a Google](https://www.google.com/)

Si queremos que la URL sea el propio enlace debemos simplemente escribir la URL entre los simbolos menor que (<)  y mayor que (>).

<https://www.google.com/>

## Codigo

Para incluir partes de codigo no ejecutable con Markdown solo es necesario utilizar 4 espacios al inicio y escribir lo que deseas en una sola linea.

Su forma es asi:

    Codigo 

    print("Hola Mundo")

Si deseas escribir mas de una sola linea de codigo es recomendable encerrar el codigo entre tres virgulillas (~ ~ ~) por arriba y tres por abajo.

Debes tener en cuenta que hay que especificar el nombre del lenguaje de programacion que estas utilizando en el codigo.

Su forma de escribirse es asi:

\~\~\~Lenguaje de programacion

Linea de codigo 1

Linea de codigo 2

etc.

\~\~\~


~~~python
for i in range(0,10,2):
    print("Hola mundo")
~~~

Si deseas resaltar codigo dentro de una frase solo debes ubicar el codigo entre acentos graves ``.

Su forma de escribirse es asi:

Aca tenemos un texto con \`codigo en el intermedio\` para luego continuar con mas texto.

De esta manera es claro que si analizamos la llamada de la funcion `suma(4,3)`, entonces obtendremos la suma de los numeros 3 y 4.

## Anular Markdown

Markdown utiliza simbolos comunes para marcar el texto de manera muy rapido (por ejemplo #, \*\*,[ ],!). En caso de querer ignorar o anular estos simbolos para utilizarlos de manera normal, es necesario anteceder al simbolo a utilizar la barra inclinada **\**.

Su forma de escribirlo es de la siguiente manera:
    
    \*Texto

4\*3\*2=24

\*\*Hola Mundo**

## Tablas con Markdown

Para elaborar una tabla que contenga filas y columnas con Markdown es necesario separar los elementos de las columnas por medio del simbolo pipline     ( | ).

Para separar el encabezado de los datos es necesario utlizar la fila de formato -------.
Si la fila no posee dos puntos entonces los datos se escribiran justificados a la izquierda, si la fila de formato posee dos puntos al inicio y al final entonces los datos se escribiran centrados. Y finalmente si la linea de formato posee dos puntos al lado derecho entonces los datos se escribiran justificados a la derecha.

Su forma de escribirse es asi:

    |Tabla|Columna 1|Columna 2|
    |-----|:--------:|-------:|
    |datos 1| 1500| 300|
    |datos 2|Hola mundo|23|

|Tabla|Columna 1|Columna 2|
|-----|:--------:|-------:|
|datos 1| 1500| 300|
|datos 2|Hola mundo|23|

## Comandos de Jupyter

Para ejecutar una celda y seguir en la siguiente: 

    Shift + Enter  

Para ejecutar una celda y permanecer en la misma: 
    
    Ctrl + Enter

Para poner el promt ( | ) en la celda solo es necesarrio pulsar:
    
    Enter

Para quitar el promt ( | ) en la celda solo es necesarrio pulsar:
    
    Esc

Para añadir una celda arriba: 

    a  

Para añadir una celda debajo: 

    b  

Para copiar una celda:
    
    c

Para cortar una celda: 

    x  

Para pegar una celda:
    
    v

Para eliminar una celda:
    
    d + d

Para combinar dos o mas celdas:
    
    Shift + m

Para separar la celda o dividir la celda en dos, ubico el cursor en el lugar deseado y ejecuto el comando:

    Ctrl + Shift + -

Para pasar de una celda de codigo a celda Markdown:
    
    m

Para pasar de una celda de Markdown a celda de codigo:
    
    y

Recuperar celdas eliminadas o cortadas:
    
    z

## Latex



(escrito LaTeX en texto sin formato) es un sistema de composición de textos, orientado a la creación de documentos escritos que presenten una alta calidad tipográfica. Por sus características y posibilidades, es usado de forma especialmente intensa en la generación de artículos y libros científicos que incluyen, entre otros elementos, expresiones matemáticas.

LaTeX está formado por un gran conjunto de macros de TeX, escrito por Leslie Lamport en 1984, con la intención de facilitar el uso del lenguaje de composición tipográfica, creado por Donald Knuth. Es muy utilizado para la composición de artículos académicos, tesis y libros técnicos, dado que la calidad tipográfica de los documentos realizados en LaTeX, se considera adecuada a las necesidades de una editorial científica de primera línea, muchas de las cuales ya lo emplean.

LaTeX es software libre bajo licencia LPPL.
