<h3>Accediendo a los elementos de un array:</h3>

División de arreglos con NumPy<br>
La división de arreglos es el proceso de extracción de un subconjunto de una de un arreglo determinada. Puedes dividir un arreglo utilizando el operador de dos puntos (:) y especificar el inicio y el final del índice del arreglo, por ejemplo:<br>

array[(filas)from:to:salto,(columnas)from:to:salto]

```python
a=np.random.random((5,5)) 
print("matriz\n",a)
print("\n")
print("indices\n",a[:,:])
print("\n")
print("indices en filas\n",a[1:3,:])
print("\n")
print("indices en filas con saltos\n",a[0:5:2,:])#con saltos
print("\n")
print("indices en columnas\n",a[:,1:3])
print("\n")
print("indices en columnas con saltos\n",a[:,0:5:2])#con saltos
print("\n")
print("indice hasta ultima fila y columna\n",a[0:-1,0:-1])
print("\n")
print("ultima columna\n",a[:,-1])
print("\n")
print("ultima fila\n",a[-1,:])
print("\n")
print("matriz sin ultima columna\n",a[:,:-1])
print("\n")
print("matriz sin primera columna\n",a[:,1:])
print("\n")
print("indices en filas y columnas con saltos\n",a[0:5:2,0:5:2])#con saltos
print("\n")
print("indices seleccionando columna\n",a[:,[1,3,4]])#seleccionando columna
```

<h3>Modificaciones en una matriz: (transponer, cambiar tamaño, cambiar dimensiones, intercambiar ejes, unir matrices, etc)</h3>

Reshape: Consiste en reorganizar el tamaño de la matriz sin alterar la dimension de la matriz<br>
numpy.reshape(a, newshape, order='C')

```python
matriz=np.random.random((5,4))
print("matriz\n", matriz)
print('Tamaño o shape de matriz: ', matriz.shape)
print('Dimension de matriz: ', matriz.ndim)

print("\n")

matriz=np.reshape(matriz,(10,2))
print("matriz despues de reshape\n", matriz)
print('Tamaño o shape de matriz: ', matriz.shape)
print('Dimension de matriz: ', matriz.ndim)

print("\n")

vector=np.array([1,2,3,4,5,6,7]
               )
print("vector columna\n", vector)
print('Tamaño o shape de vector: ', vector.shape)
print('Dimension de vector: ', vector.ndim)
vector=np.reshape(vector,(7,1))
print("\n")
print("vector columna despues de reshape\n", vector)
print('Tamaño o shape de vector: ', vector.shape)
print('Dimension de vector: ', vector.ndim)

print("\n")

x=np.arange(6).reshape((2,3))
print("x\n",x)
print(x.shape)
print(x.ndim)

print("\n")

aplanar=x.flatten()
print("aplanar\n",aplanar)
print(aplanar.shape)
print(aplanar.ndim)

print("\n")

xtranp=x.T #transpuesta de x
print("transpuesta de x\n",xtranp)
print(xtranp.shape)
print(xtranp.ndim)
```

```python
a=np.random.random((5,4))
print("matriz\n",a)
print(a.shape)
print(a.ndim)
print("\n")
b=np.random.random((5,4))
print("matriz\n",b)
print(b.shape)
print(b.ndim)
print("\n")
print("Concatenando filas de ambas matrices\n")
c=np.concatenate((a,b),axis=0) #une filas
print("matriz\n",c)
print(c.shape)
print(c.ndim)
print("\n")
print("Concatenando columnas de ambas matrices\n")
d=np.concatenate((a,b),axis=1) #une columnas
print("matriz\n",d)
print(d.shape)
print(d.ndim)
```

Repetir elementos

```python
a=np.random.random((5,4))
print("matriz\n",a)
print(a.shape)
print(a.ndim)
print("\n")
print("Repitiendo columnas con metodo tile\n")
b=np.tile(a,3) #permite repetir solo en columnas 3 veces
print("matriz\n",b)
print(b.shape)
print(b.ndim)
print("\n")
print("Repitiendo filas o columnas con metodo repeat\n")
c=np.repeat(a,3,axis=1) #permite repetir en columnas o filas 3 veces en este caso 
print("matriz\n",c)
print(c.shape)
print(c.ndim)
```

Añadiendo y removiendo elementos

```python
a=np.random.random((5,4))
print("matriz\n",a)
print(a.shape)
print(a.ndim)
print("\n")
print("Utilizando metodo delete para borrar una fila de la matriz\n")
b=np.delete(a,[1],axis=0)#borra fila 1
print("matriz\n",b)
print(b.shape)
print(b.ndim)
print("\n")
print("Utilizando metodo delete para borrar una columna de la matriz\n")
c=np.delete(a,[3],axis=1) #borra columna 4
print("matriz\n",c)
print(c.shape)
print(c.ndim)
print("\n")
print("Utilizando metodo delete para borrar varias filas de la matriz\n")
d=np.delete(a,[0,1],axis=0) #borra fila 0 y 1
print("matriz\n",d)
print(d.shape)
print(d.ndim)
print("\n")
fila=np.array([1,2,3,4]).reshape(1,4)
print("fila\n",fila)
print(fila.shape)
print(fila.ndim)
print("\n")
columna=np.array([1,2,3,4,5]).reshape(5)
print("columna\n",columna)
print(columna.shape)
print(columna.ndim)
print("\n")
print("Utilizando metodo insert para añadir fila a una posicion de la matriz\n")
e=np.insert(a,2,fila,axis=0) #añadimos fila
print("matriz\n",e)
print(e.shape)
print(e.ndim)
print("\n")
print("Utilizando metodo insert para añadir columna a una posicion de la matriz\n")
f=np.insert(a,1,columna,axis=1)#añadimo columna
print("matriz\n",f)
print(f.shape)
print(f.ndim)
print("\n")
print("Utilizando metodo append para añadir columna al final de la matriz\n")
columna=np.array([1,2,3,4,5]).reshape(5,1)
g=np.append(a,columna,axis=1)#añadimos columna al final
print("matriz\n",g)
print(g.shape)
print(g.ndim)
print("\n")
print("Utilizando metodo append para añadir fila al final de la matriz\n")
h=np.append(a,fila,axis=0)#añadimos fila al final
print("matriz\n",h)
print(h.shape)
print(h.ndim)
```

Calcular la suma, resta, multiplicación y división de dos matrices

```python
x= np.array([(1,2,3),(3,4,5)])
print("x\n",x)
y= np.array([(1,2,3),(3,4,5)])
print("y\n",y)
print("\n")
print("suma matrices\n",x+y)
print("\n")
print("resta matrices\n",x-y)
print("\n")
print("multiplicacion elemento a elemento \n",x*y)
print("\n")
print("division elemento a elemento\n",x/y)
```

Elemento minimo, elemento maximo y suma de todos los elementos de una matriz.

```python
a= np.array([2,4,8])
print("matriz\n",a)
print("\n")
print("valor minimo de la matriz\n",a.min())
print("\n")
print("valor maximo de la matriz\n",a.max())
print("\n")
print("suma de todos los elementos\n",a.sum())
```

Calcular la raíz cuadrada y la desviación estándar

```python
a=np.array([(1,2,3),(3,4,5,)])
print("matriz\n",a)
print("\n")
print("raiz de cada elemento\n",np.sqrt(a))
print("\n")
print("desviacion estandar de todos los elementos de la matriz\n",np.std(a))
```

Multiplicacion de matrices, determinante e inversa:

```python
a=np.random.random((5,4))
print("matriz A\n",a)
print("\n")
b=np.random.random((4,3))
print("matriz B\n",b)
print("\n")
c=np.dot(a,b)
print("multiplicacion de matrices C=A*B\n",c)
d=np.ones((3,3))
print("\n")
print("matriz D\n",d)
d_1=np.linalg.det(d)
print("determinante de D",d_1)
print("\n")
f=np.identity(3)
print("\n")
print("matriz F\n",f)
h=np.linalg.inv(f)
print("\n")
print("matriz F^-1\n",h)
```

Ordenar arreglos NumPy

Puede ordenar el arreglo NumPy usando el método sort () del módulo NumPy:

La función sort () toma un eje opcional (un entero) que es -1 por defecto. El eje especifica
de acuerdo a qué eje queremos ordenar el arreglo. -1 significa que el arreglo
se ordenará según el último eje.

```python
a = np.array([16, 3, 2, 6, 8, 10, 1])
print("matriz\n",a)
print("\n")
print("Sorted array = ", np.sort(a))
```

Normalizar un arreglo:

x = (x – xmin) / (xmax – xmin)

```python
x= np.array([400, 800, 200, 700, 1000, 2000, 300])
print("matriz\n",x)

xmax = x.max()

xmin = x.min()

x = (x - xmin)/(xmax - xmin)
print("\n")
print("After normalization array x = \n", x)
```
