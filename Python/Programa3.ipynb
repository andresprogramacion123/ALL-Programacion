## Cargar los datos

```python
df = pd.read_csv(data_url)
df.head()
```

## Expectativa de Vida por Continente y Año

```python
pd.pivot_table(df,values='lifeExp',columns='continent')
```

```python
pd.pivot_table(df,index='continent',values='lifeExp',columns='year')
```

```python
pd.pivot_table(df,index='year',values='lifeExp',columns='continent')
```

```python
df.pivot_table(index='year',values='lifeExp',columns='continent') #Otra forma de hacer el de arriba
```

```python
df.pivot_table(index='year',values='lifeExp',columns='continent',aggfunc='mean') #Ojo con el aggfunc, da igual que arriba
```

## Múltiples Funciones Agregadoras

```python
df.pivot_table(index='year',values='lifeExp',columns='continent',aggfunc=['min','mean','max']) 
```

## Múltiples Columnas

```python
df.pivot_table(index='year',values='pop',columns=['continent','country'])
```

## Múltiples Variables a Resumir

```python
df.pivot_table(index='year',values=['lifeExp','gdpPercap'],columns='continent')
```

## Múltiples Variables y Filas

```python
df.pivot_table(index=['continent','country'],values=['lifeExp','gdpPercap'],columns='year')
```

## Pivot con Filtro

```python
df[df.continent=='Oceania'].pivot_table(index='year',values='lifeExp',columns='country')
```
